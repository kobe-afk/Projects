```python
##---Python script created by Ivan Lim on Mar 2020. ( As Version 6 upgrades )
##---To merge all Tracking files >> REP & STACKING... then match.
import numpy, sys , os, re, time
import numpy as np
import pandas as pd
pd.options.display.max_rows = 200           # To enable custom df view size, row
pd.options.display.max_columns = 200        # To enable custom df view size, col
pd.options.display.max_colwidth = None      # To enable custom df view for max Col width
# pd.options.mode.chained_assignment = None   # default='warn'
from openpyxl import load_workbook
from datetime import datetime
from tqdm import tqdm_notebook as tqdm      # Status bar output...  For Jupyter use only.
from tqdm import tqdm                       # Enable this for Console .exe use only.
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
excl_dirs = []                      # User defined folders to ignore in search (a List) ... [ "0" , "1" ]
ext_type = ( ".PDF" , ".pdf" )     # User defined extension type of interest (a Tuple)
##########****************************************************************************************************##########
##########****************************************************************************************************##########
rootdir = 0    ### DEFAULT = 0 , to enable current working directory search only.
# rootdir = r"C:\Users\zkob\Desktop\DOE1\DOE1a no flip\DOE1a no flip"

if ( rootdir == 0 ):
    rootdir = os.getcwd()   ;   print( "Current Working Path :" , rootdir )         # Get current path directory.
    time.sleep(1)                    # IMPORTANT..... need time delay, else too fast.
else: 
    print("Alternate Working Path :" , rootdir)         # Get current path directory.
    time.sleep(1)                    # IMPORTANT..... need time delay, else too fast.
##########****************************************************************************************************##########
def CWD_Walk_Level_v5( SOURCE_DIR , Level_No = 1 , ext_type = ".pdf" , excl_dirs=[] ):
    ''' Walk thru the Directory provided & return a list of matching files (including path)
        SOURCE_DIR = url / path to investigate.
        Level_No   = User define on how deep the recursion will go.
        ext_type   = User to prove a list of file extension type to search.
        excl_dirs  = User to define if any sub-folder to exlcude ?
        RETURN: A list of matched file-type found in a path format , A list of CWD folder name.          '''
    wlist_path = []     ;       wlist_folder = []           ## Create a List container.
    #####--- Check if this is really a path provided.
    SOURCE_DIR = SOURCE_DIR.rstrip( os.path.sep )
    assert os.path.isdir( SOURCE_DIR ), print("Error Directory path !")     # Check if this is really a path provided.
    print( "Source Directory checked & OK =" , SOURCE_DIR , "\n" )          # If can print this, means "assert" is ok.
    
    num_sep = SOURCE_DIR.count( os.path.sep )       # Count Number of Seperators in the path (This is current path count).
    
    for root, dirnames, filenames in tqdm( os.walk( SOURCE_DIR ) , "MAIN LOOP " ):     #####--- MAIN LOOP.
        #yield( root, dirnames, filenames )         # Create a generator, need to "unpack" when out of this function.
        #print( root , *dirnames , sep="\n\t\t >>> " )
        #print( "\t\t\t\t >>> Folder count =" , len( dirnames ) )
        wlist_folder.append( dirnames )             # Append found directory names (need filter later b4 return)

        dirnames[:] = [ d for d in dirnames if d not in excl_dirs ]     ##--- To remove Excluded directories (if any...)

        num_sep_this = root.count( os.path.sep )    #####--- Check how deep the recursion will go.
        if num_sep + Level_No <= num_sep_this:      # Compare against current + level user requested.
            del dirnames[ : ]                       # Delete if beyond level ...
            
        for file in ( filenames ):              #####--- Check for matching file type extension.
            filepath = ( root + os.sep + file )     # Join found root path with actual filename.
            ### Check if extension met and not a ghost file (if there is a "~" at the front, then only proceed.
            if filepath.endswith( ext_type ) and file[0] != "~":    
                wlist_path.append( filepath )       # Append only the required file extension that met.

    print( "\t > End of 'CWD_Walk_Level' function : \t\t\t\t No. of Levels investigated :" , Level_No )
    print( "\t >>> Found number of files search matched =" , len( wlist_path ) )
    print( "\t >>> Found folder names from directory initial level =" , *wlist_folder[0] , sep="\n\t\t\t\t" )
    wlist_path.sort()
    #print( "\t" , *wlist_path , sep="\n\t" )
    return( wlist_path )
##########****************************************************************************************************##########

wlist_path = CWD_Walk_Level_v5( rootdir , Level_No=5 , ext_type=ext_type , excl_dirs=excl_dirs )
wlist_path
```

```python
######################## Saves the images to a specific filepath and with a specific filename that can be specified in the global variables in the main 

def saveimage(final_image_name , given_save_location ,given_directory,given_filename):
    
    global counting
    
    counting+=1
    
    number_str = str(counting)

    zero_filled_number = number_str.zfill(3)
    
    if not os.path.exists(given_directory):
        os.mkdir(os.path.join(given_save_location, given_directory))
    
    final_image_name.save(os.path.abspath(given_directory)  + "/" + zero_filled_number + "-" + given_filename + ".png" )
```

```python
###################### Places the text onto the images (now changed to only placing the text on top of the images)

def textpos(final_image ,  parameter_type , given_text_width , given_text_height ,given_main_text , given_main_text_width , given_main_text_height):
    
    Main_text = given_main_text                       # will be the newfilename from the dataframe above

    draw = ImageDraw.Draw(final_image)

    myFont = ImageFont.truetype('arial.ttf', 36)
    
    white_color = 'rgb(255, 255, 255)'                   # white color
    
    gray_color = 'rgb(105,105,105)'                      # gray color
    
#     royal_blue_color = 'rgb(65,105,225)'                 # royal blue color

    
    if parameter_type != "BLD": 
        
        draw.text((given_text_width, given_main_text_height),Main_text,fill=gray_color,font=myFont)                    # First takes in the position (x,y) then the text to draw , then the text color and font size and type
        
        return final_image
    
    else:
        
        bottom_text = "um"                             # change this to call from dhruv

        draw.text((given_main_text_width, given_main_text_height),Main_text,fill=gray_color,font=myFont)
        
        draw.text((given_text_width+103, given_text_height + 328),bottom_text,fill=white_color,font=myFont)

        return final_image
    
```

```python
##################### Only done for the BLD images 

def mergeimage(top_img , side_img,y_dist_img):
    
    ####### Resize both images to the specified height and width
    
    top_img = top_img.resize((resize_width,resize_height))
    side_img = side_img.resize((resize_width +1 ,resize_height))
    y_dist_img = y_dist_img.resize((y_dist_width,y_dist_height))
    
    ####### Merge the images together vertically 

    merged_img = Image.new('RGBA', (side_img.width, top_img.height + side_img.height + added_height))    # create a new blank image

    merged_img.paste(top_img,(0, added_height))                                                          # Paste the top image into the new blank image called merged_img

    merged_img.paste(side_img, (-1, top_img.height + added_height))                                      # On the codes above we add in text_height just to ensure that there is enough padding between where we would place the main text and the merged images              
    
    merged_img.paste(y_dist_img, (30, top_img.height + added_height +32))                                # Paste the y distance image onto the merged image and specify the specific width and height to place in the images 
    
    return merged_img
    
```

```python
######################## Only done for the flatness and roughness images

def resize_and_superimpose_image(top_image, iso_sa_image,given_width,given_height):
    
    iso_sa_image = iso_sa_image.resize((given_width,given_height))
    
    top_image.paste(iso_sa_image,(30, 140))   #35 and 150 are the x and y position where the superimposed image is 
    
    return top_image
```

```python
# specify the regions that we want to crop on the image ( For the top and side view)

def imagecrop(main_image_name,given_points):    # given points will be the dictionary we want 
    
    Cropped_img = main_image_name.crop(( given_points["left"], given_points["top"], given_points["right"], given_points["bottom"] ))   # gives us a Cropped image already

    return Cropped_img
```

```python
########################################### Resize and paste image
flatness_superimpose_resize_width = 215    #SH,FL
flatness_superimpose_resize_height = 65

unmasked_flatness_superimpose_resize_width = 180    #SH,FL
unmasked_flatness_superimpose_resize_height = 55

roughness_superimpose_resize_width = 205    # RA
roughness_superimpose_resize_height = 75

########################################### global variables for the merge image

resize_width = 440
resize_height = 365
y_dist_width = 100
y_dist_height = 35

########################################### global variables for the textpos function

added_height  = 47                                                 # The added height is used to add padding to the image so that we can place the text above the images
text_width = 40                                                    # start the adding of text on images at this width (x position)
text_height = 195                                                  # start the adding of text on images at this height (y position)
bld_text_height = added_height + 65                                # start the adding of text on images at this height for BLD only (y position)

Main_text_width = 5                                                # positioning of the main text (text above the image) in terms of the x axis
Main_text_height = 0                                               # positioning of the main text (text above the image) in terms of the y axis


########################################### global variables for the saveimage function

save_location = os.getcwd()                                       # can be changed depending on when the user wants to save it
directory = "Final images_v2_output"                                        # can be changed depending on the users needs
csv_name = "Image Values.csv"

########################################### global variables for the get_main_image function

Zygo_width = 165                                                   # The width of the zygo symbol in the pdf file
Zygo_height = 55                                                   # The height of the zygo symbol in the pdf file

topview_points = {"left":8,                                        # The topview and sideview points will be for just the flatness and the roughness
                 "top" : 8,
                 "right" : 998 ,
                 "bottom" : 593}

sideview_points = {"left":1002 ,
                "top" :8,
                "right" : 1576,
                "bottom" : 407}

unmasked_topview_points = {"left":15,                                        # The topview and sideview points will be for just the flatness and the roughness
                 "top" : 13,
                 "right" : 890 ,
                 "bottom" : 525}


topview_points_bld = {"left":200,                                  # The topview_bld and sideview_bld points is just meant for BLD 
                 "top" : 37,
                 "right" : 780 ,
                 "bottom" : 591}

sideview_points_bld = {"left":1029 ,
                "top" :15,
                "right" : 1576,
                "bottom" : 407}

y_dist_bld = {"left":860,                                          # gets the y distance from the BLD table
              "top": 635,
              "right": 896,
              "bottom":647}

new_points_Roughness = {"left":40, "top": 45, "right": 132, "bottom":78}
new_points_Flatness = {"left":128, "top": 48, "right": 217, "bottom":75}
new_points_Flatness_unmasked = {"left":190, "top": 67, "right": 317, "bottom":105}

pytesseract_new_points_R = {"left":10, "top": 40, "right": 150, "bottom":80}
pytesseract_new_points_F = {"left":0, "top": 40, "right": 220, "bottom":110}
pytesseract_new_points_B = {"left":780,"top": 634,"right": 900 ,"bottom":650}


# global iso_image
# global sa_image

def get_main_image(pdf_file_path):     
    
    pdf_file = fitz.open(pdf_file_path)
    for page_index in range(len(pdf_file)):                          # loops through the pages of the pdf files and gets a list of images
        page = pdf_file[page_index]
        image_list = page.getImageList()
    for image in image_list: 
        if image[2] > Zygo_width and image[3] > Zygo_height:         # dimensions of the zygo symbol
            xref = image[0]
            pix = fitz.Pixmap(pdf_file,xref)
            img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    
    return(img)

############################## new function called new image

def get_new_image(image, parameter_type, new_filename):
    
    if image.size == (1451, 777):
        
        top_view = imagecrop( image , unmasked_topview_points )
        iso_image = imagecrop( image , new_points_Flatness_unmasked )
        
        superimposed_img = resize_and_superimpose_image(top_view, iso_image,unmasked_flatness_superimpose_resize_width,unmasked_flatness_superimpose_resize_height)

        extend_image = Image.new('RGBA', (superimposed_img.width, superimposed_img.height + added_height))
        extend_image.paste(top_view,(0, added_height))

        final_image = textpos(extend_image , parameter_type , text_width , text_height,new_filename,Main_text_width,Main_text_height)

        saveimage(final_image , save_location , directory , new_filename)
    
    elif parameter_type == "BLD":
        
        top_view = imagecrop( image , topview_points_bld )
        side_view = imagecrop( image , sideview_points_bld ) 
        y_dist_img = imagecrop( image , y_dist_bld )

        merged_image = mergeimage(top_view , side_view , y_dist_img)

        final_image = textpos(merged_image ,  parameter_type , text_width , bld_text_height,new_filename,Main_text_width,Main_text_height)

        saveimage(final_image , save_location , directory , new_filename)

    elif parameter_type == "Flatness":

        top_view = imagecrop( image , topview_points )
        iso_image = imagecrop( image , new_points_Flatness )
        
        superimposed_img = resize_and_superimpose_image(top_view, iso_image,flatness_superimpose_resize_width,flatness_superimpose_resize_height)

        extend_image = Image.new('RGBA', (superimposed_img.width, superimposed_img.height + added_height))
        extend_image.paste(top_view,(0, added_height))

        final_image = textpos(extend_image , parameter_type , text_width , text_height,new_filename,Main_text_width,Main_text_height)

        saveimage(final_image , save_location , directory , new_filename)
    
        
    else:

        top_view = imagecrop( image , topview_points )
        sa_image = imagecrop( image , new_points_Roughness )
        
        superimposed_img = resize_and_superimpose_image(top_view, sa_image,roughness_superimpose_resize_width,roughness_superimpose_resize_height)

        extend_image = Image.new('RGBA', (superimposed_img.width, superimposed_img.height + added_height))
        extend_image.paste(top_view,(0, added_height))

        final_image = textpos(extend_image , parameter_type , text_width , text_height,new_filename,Main_text_width,Main_text_height)

        saveimage(final_image , save_location , directory , new_filename)
    
#     return(img)
```

```python
def get_text(img,file_type,i):

    new_final_df = pd.DataFrame()

    if file_type == 'Flatness':
        n_img = img.crop(( pytesseract_new_points_F["left"], pytesseract_new_points_F["top"], pytesseract_new_points_F["right"], pytesseract_new_points_F["bottom"] ))
        text = pytesseract.image_to_string(n_img)
        # text
        iso_val = re.sub("[^\d\.\-\+]", "", text.split('\n')[0])
        pv_val = re.sub("[^\d\.\-\+]", "", text.split('\n')[1])



        temp_dict = {'Parameters' : ['ISO Value', 'PV Value'], 'Value' : [iso_val,pv_val]}



        temp_df = pd.DataFrame(temp_dict)



        temp_iso= pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df[0:1].reset_index()], axis=1)
        temp_pv = pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df[1:2].reset_index()], axis=1)



        new_final_df = new_final_df.append([temp_iso,temp_pv])
        new_final_df = new_final_df.drop(columns=['index'])



    if file_type == 'Roughness':
        n_img = img.crop(( pytesseract_new_points_R["left"], pytesseract_new_points_R["top"], pytesseract_new_points_R["right"], pytesseract_new_points_R["bottom"] ))
        text = pytesseract.image_to_string(n_img)



        val = re.sub("[^\d\.\-\+]", "", text.split('\n')[0])
        val_dict = {'Parameters' : ['Sa'], 'Value' : val}
        temp_df = pd.DataFrame(val_dict)
        final_df = pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df], axis=1)



    if file_type == 'BLD':
        n_img = img.crop(( pytesseract_new_points_B["left"], pytesseract_new_points_B["top"], pytesseract_new_points_B["right"], pytesseract_new_points_B["bottom"] ))
        text = pytesseract.image_to_string(n_img)



        val = re.sub("[^\d\.\-\+]", "", text.split('\n')[0])
        val_dict = {'Parameters' : ['Y Distance'], 'Value' : val}
        temp_df = pd.DataFrame(val_dict)
        final_df = pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df], axis=1)



    return(new_final_df)
```

```python
def get_type(test_img,i):

    new_points_R = {"left":10,"top": 40,"right": 150,"bottom":180}
    z= test_img.crop(( new_points_R["left"], new_points_R["top"], new_points_R["right"], new_points_R["bottom"] ))
    output = pytesseract.image_to_string(z)

    if ("Sz" in output) or ("$z" in output):

        temp_df = pd.DataFrame(['BLD'])
        final_df = pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df], axis=1)



        return(final_df)

    elif "ISO" in output or test_img.size == (1451,777):

        temp_df = pd.DataFrame(['Flatness'])
        final_df = pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df], axis=1)

        return(final_df)

    else:

        temp_df = pd.DataFrame(['Roughness'])
        final_df = pd.concat([df_path[i:i+1].reset_index(drop = True), temp_df], axis=1)

        return(final_df)
```
